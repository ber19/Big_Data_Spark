```scala
import org.apache.spark._
import org.apache.spark.streaming._
import org.apache.spark.streaming.StreamingContext
```

```scala
// Se crea la configuracion 
val conf = new SparkConf().setMaster("local[4]").setAppName("miApp")
                .set("spark.driver.allowMultipleContexts", "true")
// Se crea el StreamingConstext
val ssc = new StreamingContext(conf, Seconds(5))
```

```scala
// Se crea el socket
val lines = ssc.socketTextStream("localhost", 9999)
// Opcion 1
// Se crea la ventana con alcance de 10 segundos
var linesWindow = lines.window(Seconds(10))
// Opcion 1
val palabras = linesWindow.flatMap(_.split(" "))

// Opcion 2
// Directamente sin crear la ventana
// val palabras = lines.flatMap(_.split(" "))

// Opcion 1
val pares = palabras.map(w => (w,1))
val cuenta = pares.reduceByKey((a,b)=>a+b)

// Opcion 2 
// Se ejecuta la funcion reduceByKeyAndWindow
// val cuenta = pares.reduceByKeyAndWindow((a,b)=>a+b, Seconds(30), Seconds(10))

cuenta.print()
```

```scala
ssc.start()
ssc.awaitTerminationOrTimeout(45000)
```

```scala
ssc.stop()
```
